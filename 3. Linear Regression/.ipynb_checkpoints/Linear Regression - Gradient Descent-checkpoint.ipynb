```python
x = np.arange(-10, 10, 1)
f_x = x**2

plt.plot(x, f_x)
plt.show()
```

```python
x_new = 10
derivative = []
y = []
learning_rate = 0.2

for i in range(100):
    old_value = x_new
    derivative.append(old_value - learning_rate * 2 * old_value)
    x_new = old_value - learning_rate * 2 * old_value
    y.append(x_new ** 2)
    

# 정해야 하는 것 1. Learning rate 2. loop
```

```python
plt.plot(x, f_x)
plt.scatter(derivative, y)
plt.show()
```

```python
def sin_function(x):
    return x * np.sin(x**2) + 1

def derivative_f(x):
    return np.sin(x**2) + 2 * (x**2) * np.cos(x**2)
```

```python
x = np.arange(-3, 3, 0.001)
f_x = sin_function(x)

plt.plot(x, f_x)
plt.show()
```

```python
x_new = 2.5
learning_rate = 0.3
derivative= []
y = []

for i in range(10):
    old_x = x_new
    derivative.append(old_x - learning_rate * derivative_f(old_x))
    x_new = old_x - learning_rate * derivative_f(old_x)
    y.append(sin_function(x_new))
    
plt.plot(x, f_x)
plt.scatter(derivative, y)
plt.show()
```

# HYPOTHESIS AND COST FUNCTION

```python
def hypothesis_function(X, theta): #가설함수 
    return X.dot(theta)

def cost_function(h, y): #비용함수
    return (1/(2*len(y))) * np.sum((h-y)**2)
```

```python
h = hypothesis_function(x,w)
cost_function(h, y) 
```

# GRADIENT DESCENT

```python
def gradient_descent(X, y, w, alpha, iterations):
    theta = w  #최적의 w값을 theta라 함. 구해야할 값. 
    m = len(y) #data의 개수
    
    theta_list = [theta.tolist()] 
    cost = cost_function(hypothesis_function(X, theta), y)
    cost_list = [cost]

    for i in range(iterations):
        t0 = theta[0] - (alpha / m) * np.sum(np.dot(X, theta) - y) #GD가 적용된 w0
        t1 = theta[1] - (alpha / m) * np.sum((np.dot(X, theta) - y) * X[:,1]) #GD가 적용된 w1
        theta = np.array([t0, t1])  #동시에 계산한 결과가 나와야하므로 t를 이용해 새로운 결과는 나중에 적용.
        
        if i % 10== 0: #cost를 중간 과정에서 확인하기 위해
            theta_list.append(theta.tolist())
            cost = cost_function(hypothesis_function(X, theta), y)
            cost_list.append(cost)

    return theta, theta_list, cost_list
```

# DO Linear regression with GD

```python
iterations = 10000
alpha = 0.001

theta, theta_list, cost_list = gradient_descent(x, y, w, alpha, iterations) #GD를 통해 w를 구하고
cost = cost_function(hypothesis_function(x, theta), y)  

print("theta:", theta)
print('cost:', cost_function(hypothesis_function(x, theta), y))
```

```python
plt.figure(figsize=(10,5))

y_predict_step= np.dot(x, np.array(theta_list).T)

plt.plot(raw_x,y,"o", alpha=0.5)
for i in range (0,len(cost_list),100):
    plt.plot(raw_x,y_predict_step[:,i], label='Line %d'%i)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# Multivariate Linear Regression

```python
x.T
```

```python
def multi_gradient_descent(x, y, theta, alpha, m, Iterations):
    xTrans = x.transpose()
    theta_list = []
    cost_list = []
    for i in range(Iterations):
        hypothesis = np.dot(x, theta)
        loss = hypothesis - y
        
        cost = np.sum(loss ** 2) / (2*m)
        gradient = np.dot(xTrans, loss) / m  # 계산된 상황을 행렬의 연산 관점으로 생각해보면 for없이 matrix 계산으로 표현가능함.
        
        #update
        theta = theta - alpha * gradient
        if i % 250 == 0:
            theta_list.append(theta)
            cost_list.append(cost)

    return theta, np.array(theta_list), cost_list #

m, n = np.shape(x)
Iterations = 10000
alpha = 0.0005
theta = np.ones(n)

theta, theta_list, cost_list = multi_gradient_descent(x, y, theta, alpha, m, Iterations)
```

```python
plt.figure(figsize=(15,5))

predict_y = np.dot(x, theta_list.T)
plt.plot(raw_x,y,"o", alpha=0.5)
for i in range (0,len(cost_list),5):
    plt.plot(raw_x, predict_y[:, i], label='Line %d'%i)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# Performance measure

- 우리가 만든 모델이 얼마나 실제 값을 잘 대변을 하는가에 대한 것을 지표로 평가하기 위해서 사용되는 여러가지 measure들 

# Mean Absolute Error(MAE)

- 잔차의 절댓값의 Sum

```python
from sklearn.metrics import median_absolute_error

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]

median_absolute_error(y_true, y_pred)
```

# Root Mean Squared Error(RMSE)

- 잔차 제곱의 Sum의 루트

```python
from sklearn.metrics import mean_squared_error

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]

mean_squared_error(y_true, y_pred)
```

# R squared

- 0 과 1사이 숫자로 크면 클수록 높은 적합도를 지님

```python
from sklearn.metrics import r2_score

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]

r2_score(y_true, y_pred)
```

# Training & Test data set

- Training한 데이터로 다시 test를 할 경우, Training 데이터에 과도하게 fitting된 모델이 될 수 있음.
- 새로운 데이터가 출현했을 때, 기존 모델과 차이 존재
- 모델은 새로운 데이터가 처리 가능하도록 generalize되야함
- 이를 위해 test set과 training set을 분리함

# Holdout Method (Sampling)

- 데이터를 Training과 Test로 나눠서 모델을 생성하고 테스트하는 기법
- 가장 일반적인 모델 생성을 위한 데이터 랜덤 샘플링 기법
- Training과 Test를 나누는 비율은 데이터의 크기에 따라 다름
- 일반적으로 Training Data 2/3, Test Data 1/3를 활용함
