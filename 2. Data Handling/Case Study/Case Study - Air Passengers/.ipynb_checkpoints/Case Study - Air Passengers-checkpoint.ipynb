- 누적 승객수, 최고, 최소 승객수?

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("AirPassengers.csv")
df["step"] = range(len(df))
df["cum_pass"] = df["#Passengers"].cumsum()
df["cum_max"] = df["#Passengers"].cummax()
df["cum_min"] = df["#Passengers"].cummin() #step을 내려가며 누적 합, 최대, 최소 구함
df
```

```python
temp_date = df["Month"].map(lambda x: x.split("-"))
temp_date = np.array(temp_date.values.tolist())     #스트링으로 묶인 데이터를 분리하는 방법
df["year"]= temp_date[:,0]
df["month"]= temp_date[:,1]
df
```

```python
df["diff"] = df["#Passengers"].diff().fillna(0)
df["pct"] = df["#Passengers"].pct_change().fillna(0).map(lambda x: round(100*x, 2)) #변화율을 넣는 방법
df.groupby(["year"]).sum()
```
